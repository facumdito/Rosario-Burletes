```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import smtplib
from email.parser import Parser
import openai
import numpy as np

# Configuración inicial
openai.api_key = 'tu-api-key-de-openai'  # Reemplazar con tu clave real

class RubberQuotationSystem:
    def __init__(self, csv_path):
        self.df = pd.read_csv(csv_path, sep=';', encoding='latin-1')
        self.clean_data()
        self.setup_similarity_model()

    def clean_data(self):
        """Limpia y prepara los datos del CSV"""
        # Eliminar filas vacías o sin información relevante
        self.df = self.df.dropna(subset=['Código', 'Descripción'])

        # Limpiar precios (convertir comas a puntos para decimales)
        self.df['Precio'] = self.df['Precio'].str.replace(',', '.').astype(float)

        # Extraer tipos de caucho principales de los códigos
        self.df['Tipo'] = self.df['Código'].apply(lambda x: x.split()[0] if pd.notna(x) else '')

    def setup_similarity_model(self):
        """Configura el modelo de similitud para búsquedas"""
        self.vectorizer = TfidfVectorizer(stop_words='spanish')
        self.descriptions = self.df['Descripción'].fillna('') + ' ' + self.df['Notas'].fillna('')
        self.tfidf_matrix = self.vectorizer.fit_transform(self.descriptions)

    def find_similar_compounds(self, query, top_n=5):
        """Encuentra compuestos similares basados en una descripción"""
        query_vec = self.vectorizer.transform([query])
        similarities = cosine_similarity(query_vec, self.tfidf_matrix).flatten()
        best_indices = np.argsort(similarities)[-top_n:][::-1]

        results = []
        for idx in best_indices:
            if similarities[idx] > 0.1:  # Umbral mínimo de similitud
                row = self.df.iloc[idx]
                results.append({
                    'Código': row['Código'],
                    'Descripción': row['Descripción'],
                    'Precio': row['Precio'],
                    'Dureza': row['Dureza'],
                    'Similitud': f"{similarities[idx]*100:.1f}%"
                })

        return results

    def get_compound_details(self, code):
        """Obtiene detalles completos de un compuesto por su código"""
        compound = self.df[self.df['Código'] == code].iloc[0].to_dict()
        return compound

    def generate_quotation(self, requirements):
        """
        Genera una cotización basada en requerimientos específicos
        :param requirements: dict con los requerimientos (tipo, dureza, aplicación, etc.)
        """
        # Filtrar por tipo si está especificado
        filtered = self.df.copy()
        if 'tipo' in requirements:
            filtered = filtered[filtered['Tipo'].str.contains(requirements['tipo'], case=False, na=False)]

        # Filtrar por dureza si está especificado
        if 'dureza_min' in requirements:
            filtered = filtered[filtered['Dureza'] >= requirements['dureza_min']]
        if 'dureza_max' in requirements:
            filtered = filtered[filtered['Dureza'] <= requirements['dureza_max']]

        # Filtrar por aplicación si está especificado
        if 'aplicacion' in requirements:
            app_filter = filtered['Descripción'].str.contains(requirements['aplicacion'], case=False) | \
                         filtered['Notas'].str.contains(requirements['aplicacion'], case=False)
            filtered = filtered[app_filter]

        # Ordenar por precio (ascendente o descendente)
        sort_order = requirements.get('orden_precio', 'asc')
        filtered = filtered.sort_values('Precio', ascending=(sort_order == 'asc'))

        # Limitar resultados si es necesario
        max_results = requirements.get('max_resultados', 10)
        results = filtered.head(max_results).to_dict('records')

        return results

    def process_email(self, email_text):
        """Procesa un correo electrónico para extraer requerimientos"""
        # Usar GPT para extraer información estructurada del correo
        prompt = f"""
        Extrae los siguientes detalles de este correo sobre una solicitud de cotización de caucho:
        - Tipo de caucho (EPDM, NBR, SBR, etc.)
        - Rango de dureza (Shore A)
        - Aplicación específica
        - Cantidad requerida
        - Otras especificaciones técnicas

        Correo: {email_text}

        Devuelve la información en formato JSON con las claves: tipo, dureza_min, dureza_max, aplicacion, cantidad, especificaciones.
        """

        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3
        )

        try:
            requirements = eval(response.choices[0].message.content)
            return requirements
        except:
            return None

    def ask_missing_info(self, missing_fields):
        """Pregunta al usuario por información faltante"""
        print("Falta información para generar la cotización:")
        responses = {}
        for field in missing_fields:
            responses[field] = input(f"{field.capitalize()}: ")
        return responses

    def create_quotation_document(self, compounds, client_info):
        """Genera un documento de cotización formal"""
        doc = f"""
        COTIZACIÓN DE COMPUESTOS DE CAUCHO
        ----------------------------------

        Cliente: {client_info.get('nombre', 'No especificado')}
        Fecha: {pd.Timestamp.now().strftime('%Y-%m-%d')}

        Requerimientos:
        - Tipo: {client_info.get('tipo', 'No especificado')}
        - Dureza: {client_info.get('dureza_min', '?')} a {client_info.get('dureza_max', '?')} Shore A
        - Aplicación: {client_info.get('aplicacion', 'No especificada')}

        Opciones recomendadas:
        """

        for i, comp in enumerate(compounds, 1):
            doc += f"""
            {i}. {comp['Código']} - {comp['Descripción']}
                - Precio: ${comp['Precio']:,.2f}
                - Dureza: {comp['Dureza']} Shore A
                {'- Notas: ' + comp.get('Notas', '') if 'Notas' in comp else ''}
            """

        doc += """

        Notas:
        - Precios sujetos a cambio según cantidad y condiciones de pago.
        - Plazo de entrega estándar: 15 días hábiles.
        """

        return doc

# Ejemplo de uso
if __name__ == "__main__":
    # Inicializar el sistema con el archivo CSV
    quotation_system = RubberQuotationSystem('todas_gomas.csv')

    # Ejemplo 1: Búsqueda por descripción
    query = "acrilo resistente a aceites para tubos"
    print(f"\nResultados para: '{query}'")
    results = quotation_system.find_similar_compounds(query)
    for res in results:
        print(f"{res['Código']} - {res['Descripción']} (Precio: ${res['Precio']:,.2f}, Dureza: {res['Dureza']}, Similitud: {res['Similitud']})")

    # Ejemplo 2: Generar cotización con requerimientos específicos
    requirements = {
        'tipo': 'EPDM',
        'dureza_min': 60,
        'dureza_max': 70,
        'aplicacion': 'túnel',
        'orden_precio': 'asc',
        'max_resultados': 5
    }

    print("\nCotización para EPDM dureza 60-70 Shore A para túnel:")
    quotation = quotation_system.generate_quotation(requirements)
    for i, item in enumerate(quotation, 1):
        print(f"{i}. {item['Código']} - ${item['Precio']:,.2f} - Dureza: {item['Dureza']}")

    # Ejemplo 3: Procesar un correo electrónico (simulado)
    sample_email = """
    Buen día,

    Necesito cotizar un compuesto de EPDM para juntas de estanqueidad en aplicaciones automotrices.
    Requerimos una dureza entre 65-75 Shore A, preferiblemente con buena resistencia a temperaturas bajas.

    La cantidad aproximada sería de 500 kg mensuales.

    Agradeceré enviar opciones con sus respectivos precios.

    Saludos,
    Carlos Méndez
    CompraTec S.A.
    """

    print("\nProcesando correo electrónico de ejemplo...")
    requirements = quotation_system.process_email(sample_email)

    if requirements:
        print("\nRequerimientos extraídos del correo:")
        print(requirements)

        # Verificar si falta información crítica
        missing_info = []
        critical_fields = ['tipo', 'dureza_min', 'dureza_max']
        for field in critical_fields:
            if field not in requirements or not requirements[field]:
                missing_info.append(field)

        if missing_info:
            responses = quotation_system.ask_missing_info(missing_info)
            requirements.update(responses)

        # Generar cotización
        quotation = quotation_system.generate_quotation(requirements)

        # Crear documento de cotización
        client_info = {
            'nombre': 'Carlos Méndez - CompraTec S.A.',
            **requirements
        }
        quotation_doc = quotation_system.create_quotation_document(quotation, client_info)

        print("\nDocumento de cotización generado:")
        print(quotation_doc)
    else:
        print("No se pudo extraer información del correo. Por favor proporcione los detalles manualmente.")
```

```python
import pandas as pd # Necesitarás instalar pandas: pip install pandas

class CotizadorCaucho:
    def __init__(self, variables_csv='variables.csv', cotizaciones_csv='cotizaciones.csv'):
        self.variables = self._cargar_variables(variables_csv)
        self.cotizaciones = self._cargar_cotizaciones(cotizaciones_csv)

    def _cargar_variables(self, filepath):
        try:
            df = pd.read_csv(filepath)
            # Convertir a un diccionario para fácil acceso
            variables_dict = {}
            for index, row in df.iterrows():
                variables_dict[row['Variable']] = [v.strip() for v in row['Valores_Posibles'].split(',')]
            return variables_dict
        except FileNotFoundError:
            print(f"Error: Archivo de variables '{filepath}' no encontrado.")
            return {}

    def _cargar_cotizaciones(self, filepath):
        try:
            return pd.read_csv(filepath)
        except FileNotFoundError:
            print(f"Error: Archivo de cotizaciones '{filepath}' no encontrado.")
            return pd.DataFrame()

    def obtener_cotizacion(self, **kwargs):
        """
        Obtiene una cotización basada en los parámetros proporcionados.
        Los parámetros deben coincidir con los nombres de las columnas en cotizaciones.csv.
        Ej: cotizador.obtener_cotizacion(Tipo_Caucho='Natural', Grado='Grado_A', Origen='Malasia', Volumen='Bajo')
        """
        if self.cotizaciones.empty:
            return "Error: No hay datos de cotización cargados."

        # Construir una consulta basada en los kwargs
        query = True
        for key, value in kwargs.items():
            if key in self.cotizaciones.columns:
                query = query & (self.cotizaciones[key] == value)
            else:
                return f"Error: La variable '{key}' no es válida para la cotización."

        resultado = self.cotizaciones[query]

        if not resultado.empty:
            # Puedes ajustar cómo se presenta el resultado. Aquí sumamos Precio_Base y Ajuste_Volumen
            precio_final = resultado['Precio_Base_USD'].iloc[0] + resultado['Ajuste_Volumen_USD'].iloc[0]
            return f"Cotización encontrada: USD {precio_final:.2f}"
        else:
            # Aquí podrías ofrecer opciones cercanas o indicar que no se encontró una coincidencia exacta.
            return "No se encontró una cotización para los parámetros dados."

    def preguntar_parametros(self):
        """
        Función interactiva para que el usuario ingrese los parámetros.
        """
        parametros = {}
        print("\nIngrese los parámetros para la cotización:")
        for variable, valores_posibles in self.variables.items():
            while True:
                print(f"  {variable}: {', '.join(valores_posibles)}")
                valor = input(f"  Ingrese el valor para {variable}: ").strip()
                if valor in valores_posibles:
                    parametros[variable] = valor
                    break
                else:
                    print(f"Valor inválido. Por favor, elija uno de: {', '.join(valores_posibles)}")
        return parametros

# --- Uso del cotizador (parte manual) ---
if __name__ == "__main__":
    cotizador = CotizadorCaucho()

    # Ejemplo de cotización interactiva
    # if cotizador.variables and not cotizador.cotizaciones.empty:
    #     parametros_solicitados = cotizador.preguntar_parametros()
    #     if parametros_solicitados:
    #         cotizacion = cotizador.obtener_cotizacion(**parametros_solicitados)
    #         print(cotizacion)
    # else:
    #     print("No se pudo inicializar el cotizador debido a errores en los archivos CSV.")

    # Ejemplo de cotización directa (simulando entrada de un correo)
    # print("\n--- Simulación de cotización desde una entrada predefinida ---")
    # parametros_ejemplo = {
    #     'Tipo_Caucho': 'Natural',
    #     'Grado': 'Grado_A',
    #     'Origen': 'Malasia',
    #     'Volumen': 'Medio'
    # }
    # cotizacion_ejemplo = cotizador.obtener_cotizacion(**parametros_ejemplo)
    # print(cotizacion_ejemplo)

    # parametros_no_encontrados = {
    #     'Tipo_Caucho': 'Sintetico',
    #     'Grado': 'Grado_C',
    #     'Origen': 'China', # Asumiendo que 'China' no está en el CSV de cotizaciones
    #     'Volumen': 'Alto'
    # }
    # cotizacion_no_encontrada = cotizador.obtener_cotizacion(**parametros_no_encontrados)
    # print(cotizacion_no_encontrada)
```

```python
import imaplib
import email
from email.header import decode_header
import re
# import spacy # pip install spacy, python -m spacy download es_core_news_sm

# Asumimos que la clase CotizadorCaucho de la Parte 1 ya está definida
# cotizador = CotizadorCaucho() # Instancia del cotizador

class LectorCorreoCotizaciones:
    def __init__(self, imap_server, email_user, email_pass, cotizador_instancia):
        self.imap_server = imap_server
        self.email_user = email_user
        self.email_pass = email_pass
        self.cotizador = cotizador_instancia
        # self.nlp = spacy.load("es_core_news_sm") # Cargar modelo de spaCy

        # Reglas simples para extracción de parámetros (puedes hacerlas más sofisticadas)
        self.palabras_clave_parametros = {
            'Tipo_Caucho': ['tipo de caucho', 'goma', 'polimero'],
            'Grado': ['grado', 'calidad'],
            'Origen': ['origen', 'procedencia', 'pais'],
            'Volumen': ['volumen', 'cantidad', 'toneladas'] # Podrías necesitar convertir '100 toneladas' a 'Alto'
        }
        self.mapeo_volumen = {
            r'\b(bajo|pocas|pequeñas?)\b': 'Bajo',
            r'\b(medio|regular)\b': 'Medio',
            r'\b(alto|grandes?|muchas)\b': 'Alto',
            r'\b(\d+(\.\d+)?)\s*(ton|tn|toneladas?)\b': 'detectar_volumen_numerico' # Para extraer números
        }


    def conectar_imap(self):
        try:
            mail = imaplib.IMAP4_SSL(self.imap_server)
            mail.login(self.email_user, self.email_pass)
            mail.select('inbox') # O la bandeja que necesites
            print(f"Conectado a IMAP en {self.imap_server}")
            return mail
        except Exception as e:
            print(f"Error al conectar o logear en IMAP: {e}")
            return None

    def desconectar_imap(self, mail):
        if mail:
            mail.logout()
            print("Desconectado de IMAP.")

    def leer_ultimos_correos(self, num_correos=5):
        mail = self.conectar_imap()
        if not mail:
            return []

        status, email_ids = mail.search(None, 'UNSEEN') # 'ALL' para todos, 'UNSEEN' para no leídos
        email_id_list = email_ids[0].split()
        correos = []

        for e_id in email_id_list[-num_correos:]: # Procesar los últimos N correos
            status, msg_data = mail.fetch(e_id, '(RFC822)')
            for response_part in msg_data:
                if isinstance(response_part, tuple):
                    msg = email.message_from_bytes(response_part[1])
                    correos.append(msg)
            # Marcar el correo como leído (opcional, si quieres procesarlo solo una vez)
            # mail.store(e_id, '+FLAGS', '\\Seen')
        self.desconectar_imap(mail)
        return correos

    def extraer_texto_correo(self, msg):
        # Decodificar el asunto
        subject_parts = decode_header(msg['Subject'])
        subject = ''
        for part, charset in subject_parts:
            if isinstance(part, bytes):
                subject += part.decode(charset if charset else 'utf-8', errors='ignore')
            else:
                subject += part

        # Extraer el cuerpo del correo
        body = ""
        if msg.is_multipart():
            for part in msg.walk():
                ctype = part.get_content_type()
                cdisp = str(part.get('Content-Disposition'))

                # Buscar texto plano
                if ctype == 'text/plain' and 'attachment' not in cdisp:
                    try:
                        charset = part.get_content_charset()
                        body = part.get_payload(decode=True).decode(charset if charset else 'utf-8', errors='ignore')
                    except UnicodeDecodeError:
                        body = part.get_payload(decode=True).decode('latin-1', errors='ignore') # Intenta con latin-1 si falla
                    break
        else:
            try:
                charset = msg.get_content_charset()
                body = msg.get_payload(decode=True).decode(charset if charset else 'utf-8', errors='ignore')
            except UnicodeDecodeError:
                body = msg.get_payload(decode=True).decode('latin-1', errors='ignore') # Intenta con latin-1 si falla
        return subject, body

    def analizar_contenido_correo(self, subject, body):
        """
        Esta es la parte "IA" inicial, donde se extraen los parámetros.
        Podría ser simple con regex o más complejo con spaCy/LLM.
        """
        texto_completo = (subject + " " + body).lower()
        parametros_extraidos = {}

        # Intento 1: Reglas y palabras clave para mapear variables
        for var, keywords in self.palabras_clave_parametros.items():
            for keyword in keywords:
                if keyword in texto_completo:
                    # Aquí necesitas una lógica más sofisticada para extraer el VALOR
                    # Por ejemplo, buscar palabras clave alrededor de la variable
                    # Esto es lo más desafiante sin un modelo de ML entrenado
                    # Para empezar, podrías buscar valores específicos de tus CSVs
                    # cerca de las palabras clave.

                    # Ejemplo muy rudimentario: buscar valores posibles cerca de la palabra clave
                    for valor_posible in self.cotizador.variables.get(var, []):
                        if valor_posible.lower() in texto_completo:
                            parametros_extraidos[var] = valor_posible
                            break # Encontró un valor, pasa a la siguiente variable
                    # Lógica específica para volumen numérico
                    if var == 'Volumen' and 'Volumen' not in parametros_extraidos:
                        for pattern, mapped_value in self.mapeo_volumen.items():
                            match = re.search(pattern, texto_completo)
                            if match:
                                if mapped_value == 'detectar_volumen_numerico':
                                    try:
                                        cantidad = float(match.group(1))
                                        if cantidad < 50: # Ejemplo de umbrales
                                            parametros_extraidos[var] = 'Bajo'
                                        elif 50 <= cantidad < 200:
                                            parametros_extraidos[var] = 'Medio'
                                        else:
                                            parametros_extraidos[var] = 'Alto'
                                    except ValueError:
                                        pass # No se pudo convertir a número
                                else:
                                    parametros_extraidos[var] = mapped_value
                                break # Ya encontró un volumen, salir del bucle de patrones
                if var in parametros_extraidos:
                    break # Si ya encontró un valor para esta variable, pasa a la siguiente

        # # Intento 2: Usar spaCy para NER si los patrones son más complejos
        # doc = self.nlp(texto_completo)
        # for ent in doc.ents:
        #     # Aquí tendrías que mapear las entidades reconocidas a tus parámetros
        #     # Esto requiere entrenamiento o reglas muy específicas para tus tipos de entidades (ej. "TipoCaucho", "GradoCaucho")
        #     if ent.label_ == "PRODUCT" and "caucho" in ent.text.lower():
        #         # Lógica para extraer tipo y grado de la entidad producto
        #         pass
        #     # ... y así sucesivamente para otras entidades relevantes

        return parametros_extraidos

    def procesar_correos(self):
        correos = self.leer_ultimos_correos(num_correos=5) # Lee los últimos 5
        for msg in correos:
            from_addr = msg['From']
            subject, body = self.extraer_texto_correo(msg)
            print(f"\n--- Procesando correo de: {from_addr} - Asunto: {subject} ---")

            parametros = self.analizar_contenido_correo(subject, body)
            print(f"Parámetros extraídos: {parametros}")

            # Identificar datos faltantes
            parametros_requeridos = list(self.cotizador.variables.keys()) # Obtiene todas las variables de tu CSV
            datos_faltantes = [p for p in parametros_requeridos if p not in parametros]

            if not datos_faltantes:
                print("Todos los parámetros encontrados. Generando cotización...")
                cotizacion = self.cotizador.obtener_cotizacion(**parametros)
                print(cotizacion)
                # Aquí enviarías un correo de respuesta con la cotización
                # self.enviar_correo_respuesta(from_addr, "Cotización Solicitada", cotizacion)
            else:
                print(f"Faltan los siguientes parámetros: {', '.join(datos_faltantes)}")
                # Preguntar los datos faltantes (requiere interacción, o un bot de chat/email)
                # Para automatización pura, deberías intentar inferir o usar valores por defecto.
                # O generar un correo de "solicitud de información adicional".
                mensaje_solicitud = f"Estimado/a,\n\nPara poder enviarle una cotización precisa, necesitamos más información sobre los siguientes puntos: {', '.join(datos_faltantes)}.\n\nPor favor, indíquenos los valores para continuar."
                # self.enviar_correo_respuesta(from_addr, "Solicitud de Información Adicional para Cotización", mensaje_solicitud)
                print(f"Enviando solicitud de información adicional a {from_addr} (simulado):\n{mensaje_solicitud}")


    # Función para enviar correos (requiere smtplib)
    def enviar_correo_respuesta(self, to_addr, subject, body):
        # Necesitarías configurar un servidor SMTP, usuario y contraseña para enviar correos.
        # Esto va más allá del alcance de este ejemplo, pero smtplib es la librería.
        # from_addr = self.email_user
        # msg = MIMEText(body)
        # msg['Subject'] = subject
        # msg['From'] = from_addr
        # msg['To'] = to_addr
        #
        # try:
        #     with smtplib.SMTP_SSL('smtp.your_email_provider.com', 465) as smtp:
        #         smtp.login(self.email_user, self.email_pass)
        #         smtp.send_message(msg)
        #     print(f"Correo de respuesta enviado a {to_addr}.")
        # except Exception as e:
        #     print(f"Error al enviar correo: {e}")
        pass # Placeholder

# --- Uso del lector de correos ---
if __name__ == "__main__":
    # Asegúrate de tener los CSVs en el mismo directorio o especifica la ruta
    cotizador = CotizadorCaucho()

    # Si los CSVs no se cargaron correctamente, no continuar
    if not cotizador.variables or cotizador.cotizaciones.empty:
        print("Error: El cotizador no se inicializó correctamente. Verifique los archivos CSV.")
    else:
        # Configura tus credenciales de correo (¡NO LAS PONGAS DIRECTAMENTE EN EL CÓDIGO FUENTE EN PRODUCCIÓN!)
        # Considera usar variables de entorno o un archivo de configuración seguro.
        EMAIL_USER = "tu_correo@ejemplo.com"
        EMAIL_PASS = "tu_contraseña" # O una contraseña de aplicación si usas 2FA
        IMAP_SERVER = "imap.ejemplo.com" # Ej: imap.gmail.com para Gmail

        # Si quieres probar el lector de correos, descomenta y rellena tus datos
        # lector = LectorCorreoCotizaciones(IMAP_SERVER, EMAIL_USER, EMAIL_PASS, cotizador)
        # lector.procesar_correos()

        print("\nPara ejecutar el lector de correos, por favor, configure sus credenciales de correo electrónico en el código.")
        print("Recuerde: para Gmail/Outlook, puede necesitar una 'contraseña de aplicación' si tiene 2FA.")
```
